```python
def fill_age(cols):
    Age=cols[0]
    Pclass=cols[1]
    if pd.isnull(Age):
        if Pclass==1:
            return 36
        if Pclass==2:
            return 29
        if Pclass==3:
            return 24
    else:
        return Age
```

```python
df['Age']=df[['Pclass','Age']].apply(fill_age,axis=1)
```
